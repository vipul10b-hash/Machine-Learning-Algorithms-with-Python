```python
# importing major libraries 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

import warnings
warnings.filterwarnings('ignore'
)
```

```python
df = pd.read_csv('Polynomial.csv').drop('Unnamed: 0',axis=1)
```

```python
x = df.iloc[:,0:1]
y = df.Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# polynomial features
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=40)
```

```python
from sklearn.linear_model import LinearRegression
lr40=LinearRegression()
```

```python
poly.fit(x_train)
#transformation
x_train_poly_40=poly.transform(x_train)
x_test_poly_40 = poly.transform(x_test)
# algorithm 
lr40 = LinearRegression()
lr40
lr40.fit(x_train_poly_40,y_train)
y_pred40 = lr40.predict(x_test_poly_40)
```

```python
x_train_poly_40[0]
```

```python
lr40.score (x_train_poly_40,y_train)
#low bias
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred40)
#high bias
```

```python
# low bias , high var ----> overfitting
# L2 --> penality ---> squaring ---> almost0
# L1 --> penality ---> absolute --> 0 --> eleminate
```

```python
from sklearn.linear_model import Ridge
r1 = Ridge()
```

```python
r1.fit(x_train_poly_40,y_train)
```

```python
r1.coef_
```

```python
# lr40.coef_
r1.score(x_train_poly_40,y_train)
```

```python
y_predr1 = r1.predict(x_test_poly_40)
```

```python
r2_score(y_test,y_predr1)
```
