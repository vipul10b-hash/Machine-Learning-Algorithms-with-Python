```python
# muliple 

# numerical features
# categorical features

# column tranformers
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python

cat = Pipeline([
    ('cat1',SimpleImputer(strategy='constant',missing_values='missing')),
    ('cat2',OneHotEncoder(sparse_output=False,dtype='int',categories=[['Female','Male']],drop='first')),
    ('cat3',OrdinalEncoder(categories=[['ssc','hsc','ug','pg']],dtype='int'))
])
cat
```

```python
num=Pipeline([
    ('num1',SimpleImputer(strategy='median')),
    ('num2',StandardScaler()),
])
num
```

```python
col1 = ColumnTransformer([
    ('cat',cat,['gender','education']),
    ('num',num,['age','salary'])
])
```

```python
from sklearn.pipeline import make_pipeline

make_pipeline(col1,LogisticRegression()
             )
```
